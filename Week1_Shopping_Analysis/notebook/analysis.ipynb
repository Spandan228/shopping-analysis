## Step 3: Data Cleaning

```python
df[['initial_price', 'final_price']].dtypes
```

```python
df['final_price'] = df['final_price'].str.replace(r'[^\d.]', '', regex=True)
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
df[['initial_price', 'final_price']].dtypes
```

```python
df = df.fillna({
    'discount': 0,
    'seller_name': 'Unknown',
    'seller_information': 'Unknown',
    'what_customers_said': 'No feedback',
    'videos': 'No video',
    'variations': 'No variations'
})
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
print("Remaining rows:", df.shape[0])
print("Remaining nulls:", df.isnull().sum().sum())
```

## Step 4: Feature Engineering

```python
df['price_difference'] = df['initial_price'] - df['final_price']
df['popularity'] = df['rating'] * df['ratings_count']
```

```python
df[['amount_of_stars', 'what_customers_said', 'seller_information', 'variations']].head()
```

```python
df['5_star_count'] = df['amount_of_stars'].str.extract(r'"5_stars":(\d+)').astype(float).fillna(0)
```

```python
df[['title', 'initial_price', 'final_price', 'price_difference', 'rating', 'ratings_count', 'popularity', '5_star_count']].head()
```

## Step 5: Analysis

### 1. Univariate Analysis

```python
plt.figure(figsize=(10, 4))
sns.histplot(df['final_price'], kde=True, bins=30)
plt.title('Distribution of Final Price')
plt.show()

df['final_price'].describe()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(df['rating'], kde=True, bins=20, color='orange')
plt.title('Distribution of Ratings')
plt.show()

df['rating'].describe()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(y=df['category'], order=df['category'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Product Categories by Count')
plt.show()

df['category'].value_counts().head(10)
```

### 2. Bivariate Analysis

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['rating'], y=df['final_price'], alpha=0.6)
plt.title('Final Price vs. Rating')
plt.show()

df[['final_price', 'rating']].corr()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['rating'], y=df['popularity'], alpha=0.6, color='purple')
plt.title('Popularity vs. Rating')
plt.show()

df[['popularity', 'rating']].corr()
```

### 3. Category-Level Analysis

```python
category_summary = df.groupby('category').agg(
    avg_price=('final_price', 'mean'),
    avg_rating=('rating', 'mean'),
    product_count=('product_id', 'count')
).reset_index()

# Display top 10 most expensive categories on average
category_summary.sort_values(by='avg_price', ascending=False).head(10)
```

```python
# Display top 10 best rated categories on average
category_summary.sort_values(by='avg_rating', ascending=False).head(10)
```

## Step 6: Visualization

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['final_price'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Final Price')
plt.xlabel('Final Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['rating'], y=df['final_price'], alpha=0.6, color='coral')
plt.title('Relationship Between Final Price and Rating')
plt.xlabel('Rating')
plt.ylabel('Final Price')
plt.show()
```

## Step 7: Insights

- **Price :** More than 80% of products are priced below 3,000, meaning economy shopping constitutes the primary demand. You should make this range your priority for any promotional campaigns, as these are the items that mainstream customers will tend to spend their money on.

- **High baseline for Customer Satisfaction:** Most Product ratings are clustered in the highly positive 4.0 to 4.7 range. And because customer satisfaction is already so high, overall, the highest area of improvement opportunity should be on increasing customer retention and repeat purchases rather than addressing quality control.

- **Final price does not determine quality/satisfaction:** There is no correlation between final_price and rating. Higher prices do not necessarily receive better reviews, showing customers expect similar or greater value for expensive products.

- **Product Popularity is Owned by Volume:** The compound popularity metric demonstrates through tabulating the number of reviews that items gain popularity mainly from sheer volume rather than statistically-devoted higher scores. More customers reviewing products is the best way to increase overall product visibility.

- **Category-Level Opportunities:** Specific categories show high average ratings despite having lower overall sales volumes. Scaling up inventory and marketing for these highly rated, low-visibility categories presents low-risk growth opportunities.

- **Inventory & Category Allocation:** The dataset is heavily concentrated in specific categories like shirts and tops. A diverse pricing strategy (offering both discount-tier and mid-tier items) within these highly populated categories is necessary to avoid customer fatigue.
